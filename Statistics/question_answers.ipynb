## Question 1
### What is the meaning of six sigma in statistics?  Give proper example

## Answer
In statistics, Sigma (Ïƒ) is a measure of variability or standard deviation. In the context of Six Sigma, the goal is to reduce the standard deviation of a process so that it operates within certain specifications, leading to fewer defects.          
Six Sigma:   
- Define: First, we define what constitutes a defect.
- Measure: we measure the current process performance.
- Analyze: we analyze the data to identify factors contributing to defects and variations.
- Improve: Based on our analysis, we make improvements to the process. 
- Control: we implement controls to ensure that the improvements are sustained over time.         
                       
For Exapmle :     
* manufacturing light bulbs, we want to ensure that they last a certain number of hours before burning out. we measure the lifespan of the bulbs and find that the average lifespan is 1000 hours with a standard deviation of 50 hours.    
                    
After implementing Six Sigma, we aim to reduce the standard deviation of the process so that the bulbs consistently last close to 1000 hours. Achieving Six Sigma quality would mean that the process produces less than 3.4 defective bulbs per million produced.

## Question 3 :
What is the meaning of the five-number summary in Statistics? Give proper example
## Answer :
The five-number summary is a descriptive statistics technique used to summarize the distribution of a dataset. It consists of five key values that divide the dataset into four equally sized parts, providing insights into the center, spread, and skewness of the data. These five numbers are:

- Minimum: The smallest value in the dataset.
- First quartile (Q1): The value below which 25% of the data falls.
- Median (Q2 or the second quartile): The middle value of the dataset when arranged in ascending order. It divides the data into two equal halves, with 50% of the data falling below and 50% above this value.
- Third quartile (Q3): The value below which 75% of the data falls.
- Maximum: The largest value in the dataset.                                


**For Example** we can take some random number like :        
{55, 60, 65, 70, 75, 80, 85, 90, 95, 100}


- Minimum: 55
- First quartile (Q1): 65 (25% of the data falls below 65)
- Median (Q2): 75 (50% of the data falls below 75)
- Third quartile (Q3): 90 (75% of the data falls below 90)
- Maximum: 100


## Question 4 :
What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook
## Answer :
Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. It quantifies the extent to which changes in one variable are associated with changes in another variable. Correlation values range from -1 to 1, where: Correlation measures the degree of relationship between two variables
           
**For Example**

```python
import numpy as np
import matplotlib.pyplot as plt
X = [5, 7, 10, 3, 8, 6, 9, 12] # Hours studied per week
Y = [70, 85, 90, 60, 80, 75, 95, 100] #Exam scores

X = np.array(X)
Y = np.array(Y)

correlation_coefficient = np.corrcoef(X, Y)[0, 1]

print("Correlation coefficient:", correlation_coefficient)

plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue')
plt.title('Scatter Plot of Hours Studied vs. Exam Scores')
plt.xlabel('Hours studied per week')
plt.ylabel('Exam scores')
plt.grid(True)
plt.show()
```
